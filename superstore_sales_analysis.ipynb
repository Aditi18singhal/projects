```python
# sale trend analysis
```

```python
# need to find yearly sales
```

```python
# convert order date(object) to date time format
```

```python
df.dtypes
```

```python
df['Order Date']=pd.to_datetime(df['Order Date'],dayfirst=True)
```

```python
# grouping by year and summing the sales per year
```

```python
yearly_sales=df.groupby(df['Order Date'].dt.year)['Sales'].sum()
yearly_sales
```

```python
plt.bar(yearly_sales.index,yearly_sales.values,color='blue')
plt.xlabel('Sales')
plt.ylabel('Year')
plt.title('Sales per year')
plt.xticks(rotation=45)
plt.show()
```
