```python
# we want to find category-wise sales
```

```python
a=df.pivot_table(index='Category',values='Total Sales (INR)',aggfunc=sum)
```

```python
a.reset_index()
```

```python
# Highest sales comes from Electronics category
```

```python
plt.bar(x=a.index,height=a['Total Sales (INR)'],color='pink')
plt.xlabel('Category')
plt.ylabel('sales')
plt.title('Distribution of sales by category')
plt.xticks(rotation=45)
plt.show()
```
