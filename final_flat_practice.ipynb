```python
#sqm sqyrd rood acre ground      #change the type to sqft
# Resale                         #Nan
# 'Ready to Move'                #Nan
```

```python
df['square_feet']=df['square_feet'].str.replace(',','')
out=[]
for i in df['square_feet']:
  a=i.split()
  if a[-1]=='sqft':
    out=out+[float(a[0])]
  elif a[-1]=='sqm':
    out=out+[float(a[0])*10.764]
  elif a[-1]=='road':
    out=out+[float(a[0])*10890]
  elif a[-1]=='acre':
    out=out+[float(a[0])*43560]
  elif a[-1]=='ground':
     out=out+[float(a[0])*2400.35]
  elif a[-1]=='sqyrd':
    out=out+[float(a[0])*9]
  elif a[-1]=='Resale':
    out=out+[np.nan]
  elif a[-1]=='Move':
    out=out+[np.nan]
  else:
    out=out+[np.nan]
df['square_feet']=out

```

```python
 #df['floor']=df['floor'].str.replace(',','/')
out=[]
for i in df['floor']:
  if pd.isna(i):
    out=out+[np.nan]
    continue
  a=i.split()
  if a[0]=='Unfurnished':
    out=out+[np.nan]
  elif a[-1]=='Property':
      out=out+[np.nan]
  elif a[0]=='Resale':
      out=out+[np.nan]
  elif a[0]=='Freehold':
      out=out+[np.nan]
  elif a[0]=='Congo':
      out=out+[np.nan]
  elif a[0]=='Ground':
      out=out+[np.nan]
  elif a[-1]=='Regalia':
      out=out+[np.nan]
  elif a[-1]=='Avenue':
      out=out+[np.nan]
  elif a[-1]=='Society':
      out=out+[np.nan]
  elif a[-1]=='Residency':
      out=out+[np.nan]
  elif a[-1]=='Furnished':
      out=out+[np.nan]
  else:
    out=out+[i]
#df['floor']=out
```

```python
l=['Unfurnished','New Property','Resale','Freehold','Congo','Ground','Milestone Regalia','The Polaris Avenue','Co-operative Society','Om Residency','Semi-Furnished']
out=[]
for i in df['floor']:
  if i in l:
    out=out+[np.nan]
  elif i!='':
    a=str(i).split(' out of ')
    out=out+[(a[0],a[-1])]
  else:
    out=out+[np.nan]
df['floor']=out

```
